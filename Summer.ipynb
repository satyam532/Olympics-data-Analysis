# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
CITY=df['City'].unique()
print(len(CITY))
    
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
DATA=df[df['Medal']=='Gold']
Smg=pd.DataFrame(DATA)
data=[]
for Medal in Smg['Sport'].unique():
    data.append([Medal, len(Smg[Smg['Sport'] ==Medal])])
pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head().plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize =(5,5))    
    
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for Medal in df['Sport'].unique():
    data.append([Medal , len(df[df['Sport']  == Medal])])
pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize =(5,5))    
    

```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for Medal in df['Athlete'].unique():
    data.append([Medal , len(df[df['Athlete']  == Medal])])
    
pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize =(5,5))      
    
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
DATA=df[df['Medal']=='Gold']
Smg=pd.DataFrame(DATA)
data = []

for Medal in Smg['Athlete'].unique():
    data.append([Medal , len(Smg[Smg['Athlete']  == Medal])])
    
pd.DataFrame(data,columns = ['Athlete','Gold_Medal']).sort_values(by='Gold_Medal', ascending=False).head().plot(x = 'Athlete', y = 'Gold_Medal', kind = 'bar', figsize =(5,5))  
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])
    

pd.DataFrame(data,columns = ['Event','Popular']).sort_values(by='Popular', ascending=False).head().plot(x = 'Event', y = 'Popular', kind = 'bar', figsize =(5,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
DATA=df[df['Medal']=='Gold']
Smg=pd.DataFrame(DATA)
Female=Smg[Smg['Gender']=='Women']
fem=pd.DataFrame(Female)
data=[]

for Gender in fem['Sport'].unique():
    data.append([Gender, len(fem[fem['Sport'] == Gender])])
    
pd.DataFrame(data,columns = ['Sport','Female_Gold']).sort_values(by='Female_Gold', ascending=False).head().plot(x = 'Sport', y = 'Female_Gold', kind = 'bar', figsize =(5,5))   
```
